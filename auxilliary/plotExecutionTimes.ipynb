```python
import re
import numpy as np
import matplotlib.pyplot as plt

numBins=50
```

```python
def findTimes(fileName1, fileName2, keyword):
    fileNames = [fileName1, fileName2]
    timePairs = []
    for fileName in fileNames:
        times=[]
        with open(fileName,"r") as inFile:
            for line in inFile.readlines():
                line = line.lstrip()
                if re.search(r"{}".format(keyword), line):
                    times.append(int(line.split("ms")[0].split(" ")[-1]))
        timePairs.append(np.array(times))
    fig,ax=plt.subplots(figsize=(14,6))
    binedges = np.histogram(np.concatenate(timePairs),bins=numBins)[1]
    plt.hist(timePairs[0],bins=binedges,alpha=0.4,density=True,label=fileName1)
    plt.hist(timePairs[1],bins=binedges,alpha=0.4,density=True,label=fileName2)
    plt.legend()
    mean1 = timePairs[0].mean()
    mean2 = timePairs[1].mean()
    plt.axvline(mean1,c='blue')
    plt.axvline(mean2,c='orange')
```

```python
fileName1="/d/grid15/ln16/pi0eta/q-values/logs/bcal/processLog0.txt"
fileName2="/d/grid15/ln16/pi0eta/q-values/logs_beforeOverhaul/bcal/processLog0.txt"
keyword="^Found neighbors"
findTimes(fileName1,fileName2,keyword)
```
